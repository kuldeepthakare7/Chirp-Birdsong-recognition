```python
mel_signal = librosa.feature.melspectrogram(y=y1, sr=sr1, n_fft=1024, hop_length=64)
mel_spect = librosa.power_to_db(np.abs(mel_signal), ref=np.max)
plt.figure(figsize=(7, 6))
librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time');
plt.title('Mel Spectrogram');
plt.colorbar(format='%+2.0f dB');
```

```python
# Compute the chroma features
chroma1 = librosa.feature.chroma_stft(y=y1, sr=sr1)

# Plot the chroma features
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma1, sr=sr1, hop_length=512, x_axis="time", y_axis="chroma")
plt.title("Chroma features")
plt.tight_layout()
plt.show()

```
